**Parameter** → variable inside the function

**Argument** → actual value passed to the function

| Type            | Description         | Example               |
| --------------- | ------------------- | --------------------- |
| Positional      | Based on position   | `add(5,3)`            |
| Keyword         | Uses parameter name | `add(a=5,b=3)`        |
| Default         | Uses default value  | `greet(name="Guest")` |


```python
# *args - Allows us to pass a variable number of non-keyword argument to a function

def multiply(*args):
  product = 1

  for i in args:
    product = product * i

  print(args)
  return(product)

multiply(1,2,3,4,5)
```

```python
# **kwargs - Allows us to pass any number of keyword argument
# keywoard arguments mean that they contain a key-value pair, like a python
# dictionary.

def display(**kwargs):

  for (key,value) in kwargs.items():
    print(key,"->",value)
```

```python
display(india="delhi",srilanka="colombo",nepal="kathmandu",pakistan="islamabad")
```
